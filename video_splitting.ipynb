Concurrent Potrait YouTube Splitter - Maximize To Fit

```python
import os
import subprocess
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.editor import *
from concurrent.futures import ThreadPoolExecutor

def download_youtube_video(url, output_path):
    # Download YouTube video using yt-dlp
    command = f'yt-dlp -o "{output_path}.%(ext)s" --format mp4 {url}'
    subprocess.call(command, shell=True)

def cut_video(input_path, output_path, start_time, end_time):
    try:
        # Cut video using moviepy
        video_clip = VideoFileClip(input_path)
        cut_clip = video_clip.subclip(start_time, end_time)
        cut_clip.write_videofile(output_path, codec='libx264')
        cut_clip.close()  # Close the cut clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while cutting the video: {e}")

def split_video_parts(video_path, output_folder):
    try:
        # Get the duration of the video
        video_clip = VideoFileClip(video_path)
        video_duration = video_clip.duration
        video_clip.close()

        # Calculate number of parts needed
        num_parts = int(video_duration) // 60 + (1 if video_duration % 60 > 25 else 0)

        # Concurrently split the video into parts
        with ThreadPoolExecutor() as executor:
            for i in range(num_parts):
                start_time = i * 60
                end_time = min((i + 1) * 60, video_duration)
                part_output_path = os.path.join(output_folder, f"Part_{i + 1}.mp4")
                executor.submit(cut_video, video_path, part_output_path, start_time, end_time)
    except Exception as e:
        print(f"Error occurred while splitting the video: {e}")

def resize_and_crop_video(input_path, output_path, width, height):
    try:
        video_clip = VideoFileClip(input_path)
        resized_clip = video_clip.resize((width, height))
        cropped_clip = resized_clip.crop(x1=(resized_clip.w - width) / 2, y1=0, x2=width, y2=height)
        cropped_clip.write_videofile(output_path, codec='libx264')
        cropped_clip.close()
        resized_clip.close()
        video_clip.close()
    except Exception as e:
        print(f"Error occurred while resizing and cropping the video: {e}")

def main():
    # Input YouTube URL
    youtube_url = input("Enter YouTube video URL: ")

    # Set output directory
    output_directory = r"D:\Splits"

    # Download the video
    video_filename = os.path.join(output_directory, "video")
    download_youtube_video(youtube_url, video_filename)

    # Check if the download was successful
    downloaded_video_path = video_filename + ".mp4"
    if not os.path.exists(downloaded_video_path):
        print("Download failed or file does not exist.")
        return

    # Get the YouTube video title using yt-dlp
    title_command = f'yt-dlp -e {youtube_url}'
    title_process = subprocess.Popen(title_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    title_output, _ = title_process.communicate()
    video_title = title_output.decode().strip()

    # Replace characters not allowed in filenames
    video_title = "".join(c if c.isalnum() or c in ['-', '_', ' '] else '_' for c in video_title)

    # Set the start time to 3 seconds
    start_time = '00:00:03'

    # Get the duration of the video
    video_clip = VideoFileClip(downloaded_video_path)
    video_duration = video_clip.duration
    video_clip.close()

    # Calculate the end time to be before the last 6 seconds
    end_time = f'00:{int(video_duration - 6) // 60}:{int(video_duration - 6) % 60}'

    # Cut the video
    cut_video_path = video_filename + "_cut.mp4"
    cut_video(downloaded_video_path, cut_video_path, start_time, end_time)

    # Check if the cut was successful
    if not os.path.exists(cut_video_path):
        print("Cutting failed or cut file does not exist.")
        return

    # Resize and crop the video to portrait format
    resized_video_path = os.path.join(output_directory, "resized_video.mp4")
    resize_and_crop_video(cut_video_path, resized_video_path, 720, 1280)

    # Check if the resize was successful
    if not os.path.exists(resized_video_path):
        print("Resizing failed or resized file does not exist.")
        return

    # Split the resized video into parts
    output_folder = os.path.join(output_directory, video_title)
    os.makedirs(output_folder, exist_ok=True)
    split_video_parts(resized_video_path, output_folder)

    # Remove the downloaded video and resized video
    os.remove(downloaded_video_path)
    os.remove(resized_video_path)

    print("Video has been successfully cut, resized, split, and exported.")

if __name__ == "__main__":
    main()
```

#### Video Splitter (Portrait Screen - Video Fit Width Only)

```python
import os
import subprocess
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.editor import *
from concurrent.futures import ThreadPoolExecutor

def download_youtube_video(url, output_path):
    # Download YouTube video using yt-dlp
    command = f'yt-dlp -o "{output_path}.%(ext)s" --format mp4 {url}'
    subprocess.call(command, shell=True)

def cut_video(input_path, output_path, start_time, end_time):
    try:
        # Cut video using moviepy
        video_clip = VideoFileClip(input_path)
        cut_clip = video_clip.subclip(start_time, end_time)
        cut_clip.write_videofile(output_path, codec='libx264')
        cut_clip.close()  # Close the cut clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while cutting the video: {e}")

def split_video_parts(video_path, output_folder):
    try:
        # Get the duration of the video
        video_clip = VideoFileClip(video_path)
        video_duration = video_clip.duration
        video_clip.close()

        # Calculate number of parts needed
        num_parts = int(video_duration) // 60 + (1 if video_duration % 60 > 25 else 0)

        # Concurrently split the video into parts
        with ThreadPoolExecutor() as executor:
            for i in range(num_parts):
                start_time = i * 60
                end_time = min((i + 1) * 60, video_duration)
                part_output_path = os.path.join(output_folder, f"Part_{i + 1}.mp4")
                executor.submit(cut_video, video_path, part_output_path, start_time, end_time)
    except Exception as e:
        print(f"Error occurred while splitting the video: {e}")

def resize_and_add_bars_video(input_path, output_path, target_width, target_height):
    try:
        video_clip = VideoFileClip(input_path)
        original_height = video_clip.h
        
        # Resize video to fit the target width
        resized_clip = video_clip.resize(width=target_width)

        # Since we are maintaining the original height, calculate padding to add to the top and bottom
        top_padding = int((target_height - original_height) / 2)
        bottom_padding = target_height - original_height - top_padding

        # Add black bars to the top and bottom
        padded_clip = CompositeVideoClip([resized_clip.set_position(("center", "center"))],
                                         size=(target_width, target_height),
                                         bg_color=(0, 0, 0))

        # Write the final video file
        padded_clip.write_videofile(output_path, codec='libx264')
        padded_clip.close()
        resized_clip.close()
        video_clip.close()
    except Exception as e:
        print(f"Error occurred while resizing and adding bars to the video: {e}")

def main():
    # Input YouTube URL
    youtube_url = input("Enter YouTube video URL: ")

    # Set output directory
    output_directory = r"D:\archive"

    # Download the video
    video_filename = os.path.join(output_directory, "video")
    download_youtube_video(youtube_url, video_filename)

    # Check if the download was successful
    downloaded_video_path = video_filename + ".mp4"
    if not os.path.exists(downloaded_video_path):
        print("Download failed or file does not exist.")
        return

    # Get the YouTube video title using yt-dlp
    title_command = f'yt-dlp -e {youtube_url}'
    title_process = subprocess.Popen(title_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    title_output, _ = title_process.communicate()
    video_title = title_output.decode().strip()

    # Replace characters not allowed in filenames
    video_title = "".join(c if c.isalnum() or c in ['-', '_', ' '] else '_' for c in video_title)

    # Set the start time to 3 seconds
    start_time = '00:00:03'

    # Get the duration of the video
    video_clip = VideoFileClip(downloaded_video_path)
    video_duration = video_clip.duration
    video_clip.close()

    # Calculate the end time to be before the last 6 seconds
    end_time = f'00:{int(video_duration - 6) // 60}:{int(video_duration - 6) % 60}'

    # Cut the video
    cut_video_path = video_filename + "_cut.mp4"
    cut_video(downloaded_video_path, cut_video_path, start_time, end_time)

    # Check if the cut was successful
    if not os.path.exists(cut_video_path):
        print("Cutting failed or cut file does not exist.")
        return

    # Resize and add black bars to the video to fit the portrait format
    resized_video_path = os.path.join(output_directory, "resized_video.mp4")
    resize_and_add_bars_video(cut_video_path, resized_video_path, 720, 1280)  # Adjust target height if needed

    # Check if the resize was successful
    if not os.path.exists(resized_video_path):
        print("Resizing failed or resized file does not exist.")
        return

    # Split the resized video into parts
    output_folder = os.path.join(output_directory, video_title)
    os.makedirs(output_folder, exist_ok=True)
    split_video_parts(resized_video_path, output_folder)

    # Remove the downloaded video and resized video
    os.remove(downloaded_video_path)
    os.remove(resized_video_path)

    print("Video has been successfully cut, resized, split, and exported.")

if __name__ == "__main__":
    main()
```

```python
import os
import subprocess
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.editor import TextClip, CompositeVideoClip, ImageClip, concatenate_videoclips
from PIL import Image

# Ensure the path to ImageMagick is set correctly in the notebook environment
os.environ["IMAGE_MAGICK_BINARY"] = r"C:\Program Files\ImageMagick-7.1.1-Q16-HDRI\magick.exe"
print(f"ImageMagick binary set to: {os.environ['IMAGE_MAGICK_BINARY']}")

# Update MoviePy configuration
change_settings({"IMAGEMAGICK_BINARY": os.environ["IMAGE_MAGICK_BINARY"]})

# Verify ImageMagick path
def verify_imagemagick():
    try:
        result = subprocess.run([os.environ["IMAGE_MAGICK_BINARY"], '--version'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        if result.returncode == 0:
            print("ImageMagick is installed and working correctly.")
        else:
            print(f"ImageMagick verification failed: {result.stderr.decode()}")
    except Exception as e:
        print(f"Error verifying ImageMagick: {e}")

verify_imagemagick()

# Update yt-dlp to nightly
def update_yt_dlp():
    command = 'yt-dlp --update-to nightly'
    subprocess.run(command, shell=True)

update_yt_dlp()

def download_youtube_video(url, output_path, cookies_path):
    # Download YouTube video using yt-dlp with cookies
    command = f'yt-dlp -o "{output_path}.%(ext)s" --format mp4 --cookies {cookies_path} {url}'
    print(f"Running command: {command}")
    result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    if result.returncode != 0:
        print(f"Failed to download video: {result.stderr.decode()}")
        return False
    return True

def download_thumbnail(url, output_path):
    # Download thumbnail using yt-dlp
    command = f'yt-dlp --write-thumbnail --skip-download --output "{output_path}" {url}'
    print(f"Running command: {command}")
    result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    if result.returncode != 0:
        print(f"Failed to download thumbnail: {result.stderr.decode()}")
        return False
    return True

def cut_video(input_path, output_path, start_time, end_time):
    try:
        # Cut video using moviepy
        video_clip = VideoFileClip(input_path)
        cut_clip = video_clip.subclip(start_time, end_time)
        
        # Write the cut video file
        cut_clip.write_videofile(output_path, codec='libx264')

        cut_clip.close()  # Close the cut clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while cutting the video: {e}")

def create_animated_text(text, duration, size, font="Arial", color="white"):
    # Create a list of text clips with increasing opacity
    text_clips = [TextClip(text, fontsize=50, font=font, color=color, size=size).set_duration(duration).crossfadein(1)]
    return concatenate_videoclips(text_clips)

def add_text_overlay(input_path, output_path, title, part_number, duration=25, countdown_duration=5):
    try:
        video_clip = VideoFileClip(input_path)
        video_duration = video_clip.duration

        # Create animated text overlay
        text = f"{title} - Part {part_number}".replace('_', ' ')
        text_clip = create_animated_text(text, duration=video_duration, size=(video_clip.w, None))

        # Create countdown timer
        countdown_clips = []
        for i in range(countdown_duration, 0, -1):
            countdown_text = f"Ends in {i}/{countdown_duration} seconds"
            countdown_clip = TextClip(countdown_text, fontsize=50, font="Arial", color="white", size=(video_clip.w, None))
            countdown_clip = countdown_clip.set_duration(1).set_position(("center", "bottom")).set_start(video_duration - countdown_duration + (countdown_duration - i))
            countdown_clips.append(countdown_clip)

        # Composite the text onto the video
        final_clip = CompositeVideoClip([video_clip, text_clip.set_position(('center', 'top')).margin(top=100)] + countdown_clips)
        final_clip.write_videofile(output_path, codec='libx264', fps=24)

        text_clip.close()
        for clip in countdown_clips:
            clip.close()
        final_clip.close()  # Close the final clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while adding text overlay: {e}")

def split_video_parts(video_path, output_folder, title):
    try:
        # Get the duration of the video
        video_clip = VideoFileClip(video_path)
        video_duration = video_clip.duration
        video_clip.close()

        # Calculate number of parts needed
        num_parts = int(video_duration) // 25 + (1 if video_duration % 25 > 0 else 0)
        
        # Check if the last part is less than 25 seconds and merge if needed
        if video_duration % 25 < 25 and num_parts > 1:
            num_parts -= 1

        for i in range(num_parts):
            start_time = i * 25
            if i == num_parts - 1:  # Adjust end time for the last part if combined
                end_time = video_duration
            else:
                end_time = (i + 1) * 25

            part_output_path = os.path.join(output_folder, f"Part_{i + 1}.mp4")  # Correct part number starting from 1
            temp_cut_path = os.path.join(output_folder, f"temp_part_{i + 1}.mp4")

            # Cut video part
            cut_video(video_path, temp_cut_path, start_time, end_time)

            # Add text overlay
            add_text_overlay(temp_cut_path, part_output_path, title, i + 1)

            # Remove temp cut file
            os.remove(temp_cut_path)

    except Exception as e:
        print(f"Error occurred while splitting the video: {e}")

def resize_and_add_bars_video(input_path, output_path, target_width, target_height, top_image, bottom_image):
    try:
        video_clip = VideoFileClip(input_path)
        original_height = video_clip.h

        # Resize video to fit the target width
        resized_clip = video_clip.resize(width=target_width)
        
        # Load top and bottom images
        top_image_clip = ImageClip(top_image, duration=resized_clip.duration).resize(width=target_width)
        bottom_image_clip = ImageClip(bottom_image, duration=resized_clip.duration).resize(width=target_width)
        
        # Calculate padding for top and bottom
        text_overlay_height = 100  # Adjust this value based on the actual text overlay height
        padding = (target_height - original_height - text_overlay_height) // 2

        # Create a black background
        background = CompositeVideoClip([resized_clip.set_position(('center', 'center'))], size=(target_width, target_height), bg_color=(0, 0, 0))
        
        # Overlay the top and bottom images on the background
        background = CompositeVideoClip([
            background,
            top_image_clip.set_position(('center', text_overlay_height + 50)),  # Move the top image slightly lower
            bottom_image_clip.set_position(('center', target_height - padding))
        ])
        
        # Write the final video file
        background.write_videofile(output_path, codec='libx264', fps=24)
        
        background.close()
        resized_clip.close()
        video_clip.close()
        top_image_clip.close()
        bottom_image_clip.close()
    except Exception as e:
        print(f"Error occurred while resizing and adding bars to the video: {e}")

def main():
    # Input YouTube URL
    youtube_url = input("Enter YouTube video URL: ")

    # Set output directory
    output_directory = r"D:\archive"
    cookies_path = r"D:\python\cookies.txt"  # Path to the cookies file

    # Download the video
    video_filename = os.path.join(output_directory, "video")
    if not download_youtube_video(youtube_url, video_filename, cookies_path):
        print("Download failed or file does not exist.")
        return

    downloaded_video_path = video_filename + ".mp4"

    # Get the YouTube video title using yt-dlp
    title_command = f'yt-dlp -e {youtube_url}'
    title_process = subprocess.Popen(title_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    title_output, _ = title_process.communicate()
    video_title = title_output.decode().strip()

    # Replace characters not allowed in filenames
    video_title = "".join(c if c.isalnum() or c in ['-', '_', ' '] else '_' for c in video_title)

    # Download the thumbnail
    thumbnail_filename = os.path.join(output_directory, "thumbnail")
    if not download_thumbnail(youtube_url, thumbnail_filename):
        print("Thumbnail download failed.")
        return

    # Get the thumbnail image path
    thumbnail_image_path = thumbnail_filename + ".jpg"
    if not os.path.exists(thumbnail_image_path):
        thumbnail_image_path = thumbnail_filename + ".webp"

    if not os.path.exists(thumbnail_image_path):
        print("Thumbnail file does not exist.")
        return

    # Split the thumbnail into top and bottom images
    thumbnail_image = Image.open(thumbnail_image_path)
    width, height = thumbnail_image.size
    top_image_path = os.path.join(output_directory, "top_thumbnail.jpg")
    bottom_image_path = os.path.join(output_directory, "bottom_thumbnail.jpg")
    thumbnail_image.crop((0, 0, width, height // 2)).save(top_image_path)
    thumbnail_image.crop((0, height // 2, width, height)).save(bottom_image_path)

    # Set the start time to 3 seconds
    start_time = 3  # 3 seconds in seconds

    # Get the duration of the video
    video_clip = VideoFileClip(downloaded_video_path)
    video_duration = video_clip.duration
    video_clip.close()

    # Calculate the end time to be before the last 6 seconds
    end_time = video_duration - 6  # in seconds

    # Cut the video
    cut_video_path = video_filename + "_cut.mp4"
    cut_video(downloaded_video_path, cut_video_path, start_time, end_time)

    # Check if the cut was successful
    if not os.path.exists(cut_video_path):
        print("Cutting failed or cut file does not exist.")
        return

    # Resize and add the thumbnail images as bars to the video to fit the portrait format
    target_width, target_height = 1080, 1920  # TikTok recommended dimensions
    resized_video_path = os.path.join(output_directory, "resized_video.mp4")
    resize_and_add_bars_video(cut_video_path, resized_video_path, target_width, target_height, top_image_path, bottom_image_path)

    # Check if the resize was successful
    if not os.path.exists(resized_video_path):
        print("Resizing failed or resized file does not exist.")
        return

    # Split the resized video into parts
    output_folder = os.path.join(output_directory, video_title)
    os.makedirs(output_folder, exist_ok=True)
    split_video_parts(resized_video_path, output_folder, video_title)

    # Ensure all clips are closed
    resized_video_clip = VideoFileClip(resized_video_path)
    resized_video_clip.close()

    # Remove the downloaded video and resized video
    os.remove(downloaded_video_path)
    os.remove(resized_video_path)

    print("Video has been successfully cut, resized, split, and exported.")

if __name__ == "__main__":
    main()
```

```python
import os
import subprocess
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.editor import TextClip, CompositeVideoClip

# Ensure the path to ImageMagick is set correctly in the notebook environment
os.environ["IMAGE_MAGICK_BINARY"] = r"C:\Program Files\ImageMagick-7.1.1-Q16-HDRI\magick.exe"
print(f"ImageMagick binary set to: {os.environ['IMAGE_MAGICK_BINARY']}")

# Update MoviePy configuration
change_settings({"IMAGEMAGICK_BINARY": os.environ["IMAGE_MAGICK_BINARY"]})

# Verify ImageMagick path
def verify_imagemagick():
    try:
        result = subprocess.run([os.environ["IMAGE_MAGICK_BINARY"], '--version'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        if result.returncode == 0:
            print("ImageMagick is installed and working correctly.")
        else:
            print(f"ImageMagick verification failed: {result.stderr.decode()}")
    except Exception as e:
        print(f"Error verifying ImageMagick: {e}")

verify_imagemagick()

def download_youtube_video(url, output_path):
    # Download YouTube video using yt-dlp
    command = f'yt-dlp -o "{output_path}.%(ext)s" --format mp4 {url}'
    print(f"Running command: {command}")
    result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    if result.returncode != 0:
        print(f"Failed to download video: {result.stderr.decode()}")
        return False
    return True

def cut_video(input_path, output_path, start_time, end_time):
    try:
        # Cut video using moviepy
        video_clip = VideoFileClip(input_path)
        cut_clip = video_clip.subclip(start_time, end_time)
        
        # Write the cut video file
        cut_clip.write_videofile(output_path, codec='libx264')

        cut_clip.close()  # Close the cut clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while cutting the video: {e}")

def add_text_overlay(input_path, output_path, title, part_number):
    try:
        video_clip = VideoFileClip(input_path)

        # Add text overlay
        text = f"{title} - Part {part_number}".replace('_', ' ')
        text_clip = TextClip(
            text,
            fontsize=40,
            color='white',
            method='caption',
            size=(video_clip.w, None)  # Make the text fit the video width
        ).set_duration(video_clip.duration).set_fps(24)
        text_clip = text_clip.set_position(('center', 'top')).margin(top=70)  # Adjust the margin to move the text downwards

        # Composite the text onto the video
        final_clip = CompositeVideoClip([video_clip, text_clip])
        final_clip.write_videofile(output_path, codec='libx264', fps=24)

        text_clip.close()
        final_clip.close()  # Close the final clip
        video_clip.close()  # Close the original clip
    except Exception as e:
        print(f"Error occurred while adding text overlay: {e}")

def split_video_parts(video_path, output_folder, title):
    try:
        # Get the duration of the video
        video_clip = VideoFileClip(video_path)
        video_duration = video_clip.duration
        video_clip.close()

        # Calculate number of parts needed
        num_parts = int(video_duration) // 25 + (1 if video_duration % 25 > 0 else 0)
        
        # Check if the last part is less than 25 seconds and merge if needed
        if video_duration % 25 < 25 and num_parts > 1:
            num_parts -= 1

        for i in range(num_parts):
            start_time = i * 25
            if i == num_parts - 1:  # Adjust end time for the last part if combined
                end_time = video_duration
            else:
                end_time = (i + 1) * 25

            part_output_path = os.path.join(output_folder, f"Part_{i + 1}.mp4")  # Correct part number starting from 1
            temp_cut_path = os.path.join(output_folder, f"temp_part_{i + 1}.mp4")

            # Cut video part
            cut_video(video_path, temp_cut_path, start_time, end_time)

            # Add text overlay
            add_text_overlay(temp_cut_path, part_output_path, title, i + 1)

            # Remove temp cut file
            os.remove(temp_cut_path)

    except Exception as e:
        print(f"Error occurred while splitting the video: {e}")

def resize_and_add_bars_video(input_path, output_path, target_width, target_height):
    try:
        video_clip = VideoFileClip(input_path)
        original_height = video_clip.h
        
        # Resize video to fit the target width
        resized_clip = video_clip.resize(width=target_width)

        # Since we are maintaining the original height, calculate padding to add to the top and bottom
        top_padding = int((target_height - original_height) / 2)
        bottom_padding = target_height - original_height - top_padding

        # Add black bars to the top and bottom
        padded_clip = CompositeVideoClip([resized_clip.set_position(("center", "center"))],
                                         size=(target_width, target_height),
                                         bg_color=(0, 0, 0))

        # Write the final video file
        padded_clip.write_videofile(output_path, codec='libx264', fps=24)
        padded_clip.close()
        resized_clip.close()
        video_clip.close()
    except Exception as e:
        print(f"Error occurred while resizing and adding bars to the video: {e}")

def main():
    # Input YouTube URL
    youtube_url = input("Enter YouTube video URL: ")

    # Set output directory
    output_directory = r"D:\archive"

    # Download the video
    video_filename = os.path.join(output_directory, "video")
    if not download_youtube_video(youtube_url, video_filename):
        print("Download failed or file does not exist.")
        return

    downloaded_video_path = video_filename + ".mp4"

    # Get the YouTube video title using yt-dlp
    title_command = f'yt-dlp -e {youtube_url}'
    title_process = subprocess.Popen(title_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    title_output, _ = title_process.communicate()
    video_title = title_output.decode().strip()

    # Replace characters not allowed in filenames
    video_title = "".join(c if c.isalnum() or c in ['-', '_', ' '] else '_' for c in video_title)

    # Set the start time to 3 seconds
    start_time = 3  # 3 seconds in seconds

    # Get the duration of the video
    video_clip = VideoFileClip(downloaded_video_path)
    video_duration = video_clip.duration
    video_clip.close()

    # Calculate the end time to be before the last 6 seconds
    end_time = video_duration - 6  # in seconds

    # Cut the video
    cut_video_path = video_filename + "_cut.mp4"
    cut_video(downloaded_video_path, cut_video_path, start_time, end_time)

    # Check if the cut was successful
    if not os.path.exists(cut_video_path):
        print("Cutting failed or cut file does not exist.")
        return

    # Resize and add black bars to the video to fit the portrait format
    resized_video_path = os.path.join(output_directory, "resized_video.mp4")
    resize_and_add_bars_video(cut_video_path, resized_video_path, 720, 1280)  # Adjust target height if needed

    # Check if the resize was successful
    if not os.path.exists(resized_video_path):
        print("Resizing failed or resized file does not exist.")
        return

    # Split the resized video into parts
    output_folder = os.path.join(output_directory, video_title)
    os.makedirs(output_folder, exist_ok=True)
    split_video_parts(resized_video_path, output_folder, video_title)

    # Ensure all clips are closed
    resized_video_clip = VideoFileClip(resized_video_path)
    resized_video_clip.close()

    # Remove the downloaded video and resized video
    os.remove(downloaded_video_path)
    os.remove(resized_video_path)

    print("Video has been successfully cut, resized, split, and exported.")

if __name__ == "__main__":
    main()
```
